```python
# Import required libraries and dependencies
import pandas as pd
import numpy as np
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
ss = StandardScaler()
market_scaled = ss.fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
df_market_scaled = pd.DataFrame(market_scaled, 
                                columns=df_market_data.columns,
                                index=df_market_data.index)

# Display sample data
df_market_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
def elbow_method(data, title=None):
    # Create a list with the number of k-values from 1 to 11
    k = list(range(1,11))
    # Create an empty list to store the inertia values
    inertia = []
    
    # Create a for loop to compute the inertia with each possible value of k
    for i in k:
        # Create a KMeans model using the loop counter for the n_clusters
        k_model = KMeans(n_clusters=i, random_state=1)
        # Fit the model to the data using `df_market_data_scaled
        k_model.fit(data)
        # Append the model.inertia_ to the inertia list
        inertia.append(k_model.inertia_)
        
    # Create a DataFrame with the data to plot the Elbow curve
    elbow_df = pd.DataFrame({"k": k, "inertia": inertia})
    plot = elbow_df.hvplot.line(
                            x="k", 
                            y="inertia", 
                            title=f"Elbow Curve for {title}", 
                            xticks=k
                                )
    return plot
```

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
og_elbow = elbow_method(df_market_scaled, title="orignal data")
og_elbow
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** k=4 looks like the optimal value of k for minimizing overall inertia according to the elbow method. 

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
def K_model(data, n):
    # Initialize the K-Means model using the best value for k
    model = KMeans(n_clusters=n, random_state=1)
    # Fit the K-Means model using the scaled data
    model.fit(data)
    # Predict the clusters to group the cryptocurrencies using the scaled data
    k_preds = model.predict(data)
    # Create a copy of the DataFrame
    df_preds = data.copy()
    # Add a new column to the DataFrame with the predicted clusters
    df_preds["cluster_identity"] = k_preds
    return df_preds

df_market_clustered = K_model(df_market_scaled, 4)

# Display sample data
df_market_clustered.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
og_clusters = df_market_clustered.hvplot.scatter(
                                                    x="price_change_percentage_24h",
                                                    y="price_change_percentage_7d",
                                                    by="cluster_identity",
                                                    hover_cols = "coin_id",
                                                    title="OG Clustering"
                                                    )
og_clusters
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 
# three principal components.
pca_market_vals = pca.fit_transform(df_market_scaled)

# View the first five rows of the DataFrame. 
pca_market_vals[:5]
```

```python
# Retrieve the explained variance to determine how much information 
# can be attributed to each principal component.
print(pca.explained_variance_ratio_)
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The PCA model is accounting for just under 90% of the variance in the total distribution (~89.5%) which is remarkably strong for only having 41 objects with 7 different dimensions for those objects. 

```python
def pca_instance(data, n, original_data):
    # Create a PCA model instance and set `n_components=n`.
    pca = PCA(n_components=n)
    # Use the PCA model with `fit_transform` to reduce to n principal components.
    info = pca.fit_transform(data)
    # Create a new DataFrame with the PCA data.
    pca_df = pd.DataFrame( info,  columns=[ f"PC{i}" for i in range(1,n+1) ] )
    # Copy the crypto names from the original data
    pca_df.set_index(original_data.index, inplace=True)
    return pca_df

# Display sample data
pca_market_df = pca_instance(df_market_scaled, 3, df_market_data)
pca_market_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# Plot a line chart with all the inertia values computed with 
# the different values of k to visually identify the optimal value for k.
pca_elbow = elbow_method(pca_market_df, title="PCA data")
pca_elbow
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** k=4 looks like the optimal value of k for minimizing overall inertia according to the elbow method. 


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** This answer does not differ from the best k-value for the original data, however, the overall inertia values for the PCA data are all lower than the original data. This makes sense considering the PCA model is attempting to capture as much of the original data as possible, and reducing the dimesionality of the data will result in denser regions of data and lower inertia values across the board.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# Initialize the K-Means model using the best value for k
# Fit the K-Means model using the PCA data
# Predict the clusters to group the cryptocurrencies using the PCA data
# Create a copy of the DataFrame with the PCA data
# Add a new column to the DataFrame with the predicted clusters
# Display sample data
pca_df_market_clustered = K_model(pca_market_df,4)
pca_df_market_clustered.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PC1"` and `y="PC2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
pca_clusters = pca_df_market_clustered.hvplot.scatter(
                                                        x="PC1",
                                                        y="PC2",
                                                        by="cluster_identity",
                                                        hover_cols = "coin_id",
                                                        title="PCA Clustering"
                                                        )
pca_clusters
```
